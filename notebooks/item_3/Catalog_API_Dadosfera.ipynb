#1.   **Importações**

```python
import os
import json
import requests
import pandas as pd
from typing import Any, Dict, List, Optional
from google.colab import userdata
```

#2.  **Configuração das variáveis de ambiente**



```python
BASE_URL = "https://maestro.dadosfera.ai"
LANG = "pt-br"
TIMEOUT = 60

# Variáveis de acesso definidas como variáveis de ambiente no colab por questões de segurança
USERNAME = userdata.get('DADOSFERA_USERNAME')
PASSWORD = userdata.get('DADOSFERA_PASSWORD')
ASSET_ID = userdata.get('DADOSFERA_ASSET_ID')
```

#3.   **Primeira conexão para obter o token de acesso**



```python
resp = requests.post(
    f"{BASE_URL}/auth/sign-in",
    headers={
        "Content-Type": "application/json",
        "dadosfera-lang": "pt-br"
    },
    json={
        "username": USERNAME,
        "password": PASSWORD
    }
)

print(resp.status_code)
print(resp.text)

data = resp.json()
token = data["tokens"]["accessToken"]

print("Token OK")
```

#4.   **Listar data assets presentes no catálogo**



```python
resp = requests.get(
    f"{BASE_URL}/catalog?order=asc",
    headers={
        "Authorization": f"{token}",
        "dadosfera-lang": "pt-br"
    },
    params={
        "page": 1,
        "size": 100
    }
)

print(resp.status_code)
print(resp.text[:5000])

catalogo = resp.json()
print(json.dumps(catalogo, indent=2, ensure_ascii=False)[:10000])
```

#5.   **Listar data asset específico por meio do ID**



```python
resp = requests.get(
    f"{BASE_URL}/catalog/data-asset/{ASSET_ID}",
    headers={
        "Authorization": f"{token}",
        "dadosfera-lang": "pt-br"
    },
    params={
        "shared": "false"
    }
)

print(resp.status_code)
detalhe = resp.json()
print(json.dumps(detalhe, indent=2, ensure_ascii=False)[:10000])
```

#6.   **Adicionar informações de tags e descrição no data asset selecionado**






```python
resp = requests.put(
    f"{BASE_URL}/catalog/data-asset/{ASSET_ID}",
    headers={
        "Authorization": f"{token}",
        "Content-Type": "application/json",
        "dadosfera-lang": "pt-br"
    },
    json={
        "name": "PUBLIC.OLIST_ORDER_ITEMS_DATASET_ATHOS_JOHANN",
        "description": """Dataset contendo os itens de pedidos do e-commerce Olist.

Cada linha representa um item dentro de um pedido, incluindo produto, vendedor, preço, frete e data limite de envio.

Granularidade:
1 linha = 1 item de pedido

Organização em Data Lake:
- Bronze: dados brutos ingeridos via pipeline
- Silver: dados tratados via microtransformação
- Gold: dados refinados para consumo analítico
""",
        "tags": [
            "ecommerce",
            "olist",
            "orders",
            "items",
            "analytics",
            "bronze",
            "silver",
            "gold"
        ],
        "embed": {
            "share_type": "private"
        }
    }
)

print(resp.status_code)
detalhe = resp.json()
print(json.dumps(detalhe, indent=2, ensure_ascii=False)[:10000])
```

#7.   **Realizar preview do data asset**



```python
resp = requests.get(
    f"{BASE_URL}/catalog/data-asset/{ASSET_ID}/preview",
    headers={
        "Authorization": f"{token}",
        "dadosfera-lang": "pt-br"
    }
)

print(resp.status_code)
preview = resp.json()
print(json.dumps(preview, indent=2, ensure_ascii=False)[:10000])
```

#8.   **Verificar metadados do data asset**



```python
resp = requests.get(
    f"{BASE_URL}/catalog/data-asset/{ASSET_ID}/columns-metadata",
    headers={
        "Authorization": f"{token}",
        "dadosfera-lang": "pt-br"
    }
)

print(resp.status_code)
metadata = resp.json()
print(json.dumps(metadata, indent=2, ensure_ascii=False)[:10000])
```

#9.   **Adicionar documentação**



```python
resp = requests.post(
    f"{BASE_URL}/catalog/data-asset/{ASSET_ID}/docs",
    headers={
        "Authorization": f"{token}",
        "Content-Type": "application/json",
        "dadosfera-lang": "pt-br"
    },
    params={
        "asset_type": "data-asset"
    },
    json={
        "docs": """# Dataset: Olist Order Items

## Contexto de Negócio

Este dataset contém informações sobre os itens de pedidos do e-commerce Olist.

Cada registro representa um item dentro de um pedido, permitindo análises detalhadas sobre vendas, logística e comportamento de compra.

O dataset é fundamental para entender:
- desempenho de vendas por produto
- performance de sellers
- composição de pedidos
- impacto do frete no valor final
- eficiência logística

## Granularidade

Cada linha do dataset representa **1 item dentro de um pedido**.

Ou seja, um pedido pode possuir múltiplos registros neste dataset.

## Estrutura dos Dados

O dataset contém informações relacionadas a:
- pedidos (`ORDER_ID`)
- itens (`ORDER_ITEM_ID`)
- produtos (`PRODUCT_ID`)
- vendedores (`SELLER_ID`)
- valores financeiros (`PRICE`, `FREIGHT_VALUE`)
- logística (`SHIPPING_LIMIT_DATE`)

## Organização em Data Lake

### Bronze
Dados brutos ingeridos via pipeline, sem transformação significativa.

### Silver
Dados tratados via microtransformação, incluindo:
- padronização de colunas
- ajuste de tipos de dados
- tratamento de inconsistências

### Gold
Dados refinados e prontos para consumo analítico, como:
- vendas por produto
- vendas por seller
- ticket médio
- análise de frete

## Dicionário de Dados

- `ORDER_ID`: Identificador único do pedido
- `ORDER_ITEM_ID`: Número sequencial do item dentro do pedido
- `PRODUCT_ID`: Identificador do produto
- `SELLER_ID`: Identificador do vendedor
- `SHIPPING_LIMIT_DATE`: Data limite para envio do item
- `PRICE`: Valor do item no pedido
- `FREIGHT_VALUE`: Valor do frete associado ao item

## Possíveis Análises

Este dataset pode ser utilizado para:
- análise de vendas por produto
- análise de desempenho de vendedores
- cálculo de ticket médio
- análise de custo de frete
- identificação de padrões de compra
- análise de eficiência logística

## Governança e Uso

O dataset foi catalogado na plataforma Dadosfera com o objetivo de:
- garantir governança dos dados
- facilitar descoberta e reutilização
- padronizar documentação
- permitir integração com pipelines analíticos

## Observações Técnicas

- Dataset proveniente de carga Full Load via pipeline
- Estrutura otimizada para análises em nível de item
- Pode ser combinado com outros datasets do Olist para análises mais completas, como orders, customers e products

## Linhagem de Dados

- Fonte: Pipeline de ingestão Olist
- Processamento: Microtransformação na Dadosfera
- Destino: Data Asset catalogado para análise
"""
    }
)

print(resp.status_code)
print(resp.text)
```
