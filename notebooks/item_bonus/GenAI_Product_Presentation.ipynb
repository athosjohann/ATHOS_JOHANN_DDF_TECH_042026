````python
import os
import base64
import io
import pandas as pd
import matplotlib.pyplot as plt
from PIL import Image
from openai import OpenAI
from google.colab import userdata

client = OpenAI(api_key=userdata.get('OPENAI_API_KEY'))
````

## 2. Funções auxiliares

````python
def generate_image(prompt: str, size: str = "1024x1024") -> Image.Image:
    """Chama DALL-E 3 e retorna um objeto PIL.Image."""
    response = client.images.generate(
        model="dall-e-3",
        prompt=prompt,
        size=size,
        quality="standard",
        response_format="b64_json",
        n=1,
    )
    img_bytes = base64.b64decode(response.data[0].b64_json)
    return Image.open(io.BytesIO(img_bytes))


def plot_image(img: Image.Image, title: str = "") -> None:
    """Exibe a imagem gerada com matplotlib."""
    plt.figure(figsize=(6, 6))
    plt.imshow(img)
    plt.axis('off')
    if title:
        plt.title(title, fontsize=13, pad=10)
    plt.tight_layout()
    plt.show()
````

## 3. Prompts utilizados — Documentação

### Estratégia de prompt

O prompt segue a estrutura:
```
Professional e-commerce product image: {nome}, {categoria}.
{estilo_visual}.
High quality, commercial photography style, suitable for online marketplace.
No text or watermarks.
```

**Variáveis:**
- `nome`: nome do produto fornecido pelo usuário
- `categoria`: categoria do produto
- `estilo_visual`: um dos estilos pré-definidos no app

**Restrição:** `No text or watermarks` é essencial — DALL-E tende a adicionar texto falso em imagens de produto.

---

### Prompts testados

````python
# ── PROMPT 1 — Estilo foto comercial (fundo branco) ──────────────────────────
prompt_1 = (
    "Professional e-commerce product image: Fone de Ouvido Bluetooth Premium, Eletrônicos. "
    "foto comercial profissional, fundo branco, iluminação de estúdio. "
    "High quality, commercial photography style, suitable for online marketplace. "
    "No text or watermarks."
)

print("Prompt 1:")
print(prompt_1)
img_1 = generate_image(prompt_1)
plot_image(img_1, "Fone Bluetooth — Fundo branco")
````

````python
# ── PROMPT 2 — Estilo lifestyle ───────────────────────────────────────────────
prompt_2 = (
    "Professional e-commerce product image: Mochila Executiva Impermeável, Acessórios. "
    "lifestyle moderno, ambiente clean e minimalista. "
    "High quality, commercial photography style, suitable for online marketplace. "
    "No text or watermarks."
)

print("Prompt 2:")
print(prompt_2)
img_2 = generate_image(prompt_2)
plot_image(img_2, "Mochila — Lifestyle")
````

````python
# ── PROMPT 3 — Estilo flat lay ────────────────────────────────────────────────
prompt_3 = (
    "Professional e-commerce product image: Kit Skincare Facial, Beleza e Saúde. "
    "flat lay elegante sobre superfície de mármore. "
    "High quality, commercial photography style, suitable for online marketplace. "
    "No text or watermarks."
)

print("Prompt 3:")
print(prompt_3)
img_3 = generate_image(prompt_3)
plot_image(img_3, "Kit Skincare — Flat lay")
````

````python
# ── PROMPT 4 — Renderização 3D ────────────────────────────────────────────────
prompt_4 = (
    "Professional e-commerce product image: Smartwatch Esportivo, Eletrônicos. "
    "renderização 3D realista, fundo gradiente azul escuro. "
    "High quality, commercial photography style, suitable for online marketplace. "
    "No text or watermarks."
)

print("Prompt 4:")
print(prompt_4)
img_4 = generate_image(prompt_4)
plot_image(img_4, "Smartwatch — Renderização 3D")
````

## 4. Análise dos resultados

| Prompt | Estilo | Qualidade observada |
