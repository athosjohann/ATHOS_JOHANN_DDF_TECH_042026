```python
import pandas as pd
import great_expectations as gx
import json
```

```python
# Caminho do arquivo CSV
arquivo = "/content/olist_order_items_dataset.csv"

# Leitura do CSV
df = pd.read_csv(arquivo)

# Padroniza nomes das colunas
df.columns = [c.upper() for c in df.columns]
df.head()
```

```python
# Criar o contexto do Great Expectations
context = gx.get_context()

# Criar uma fonte de dados baseada em pandas
data_source = context.data_sources.add_pandas(name="pandas_source")

# Criar um asset
data_asset = data_source.add_dataframe_asset(name="olist_asset")
batch_definition = data_asset.add_batch_definition_whole_dataframe("full_batch")

# Criar o batch com o dataframe
batch = batch_definition.get_batch(
    batch_parameters={"dataframe": df}
)

# Criar o validator
validator = context.get_validator(
    batch=batch,
    create_expectation_suite_with_name="suite_olist"
)
```

```python
# Lista para guardar os resultados
resultados = []

# Regras de qualidade

# Não pode ser nulo
resultados.append(validator.expect_column_values_to_not_be_null("ORDER_ID"))
resultados.append(validator.expect_column_values_to_not_be_null("ORDER_ITEM_ID"))
resultados.append(validator.expect_column_values_to_not_be_null("PRODUCT_ID"))
resultados.append(validator.expect_column_values_to_not_be_null("SELLER_ID"))

# Valores válidos
resultados.append(validator.expect_column_values_to_be_between("PRICE", min_value=0.01))
resultados.append(validator.expect_column_values_to_be_between("FREIGHT_VALUE", min_value=0))

# Data não pode ser nula
resultados.append(validator.expect_column_values_to_not_be_null("SHIPPING_LIMIT_DATE"))
```

```python
# Cria uma chave única
df["ORDER_ITEM_KEY"] = df["ORDER_ID"].astype(str) + "_" + df["ORDER_ITEM_ID"].astype(str)

# Atualiza batch com nova coluna
batch = batch_definition.get_batch(
    batch_parameters={"dataframe": df}
)

validator = context.get_validator(
    batch=batch,
    expectation_suite_name="suite_olist"
)

# Verifica duplicidade
resultados.append(validator.expect_column_values_to_be_unique("ORDER_ITEM_KEY"))
```

```python
resumo = []

# Loop nos resultados para gerar um relatório simples
for r in resultados:
    resumo.append({
        "regra": r["expectation_config"]["type"],
        "coluna": r["expectation_config"]["kwargs"].get("column", ""),
        "sucesso": r["success"],
        "erros": r["result"].get("unexpected_count", 0)
    })

df_resumo = pd.DataFrame(resumo)
df_resumo
```

```python
import json

# Métricas manuais
relatorio = {
    "total_linhas": len(df),
    "nulos_order_id": int(df["ORDER_ID"].isnull().sum()),
    "nulos_product_id": int(df["PRODUCT_ID"].isnull().sum()),
    "nulos_seller_id": int(df["SELLER_ID"].isnull().sum()),
    "price_invalido": int((df["PRICE"] <= 0).sum()),
    "frete_invalido": int((df["FREIGHT_VALUE"] < 0).sum()),
    "duplicidade_item": int(df["ORDER_ITEM_KEY"].duplicated().sum())
}

# Mostra em formato JSON
print(json.dumps(relatorio, indent=2, ensure_ascii=False))
```

```python
# Salva relatório das regras
df_resumo.to_csv("/content/relatorio_quality.csv", index=False)

# Salva métricas gerais
pd.DataFrame([relatorio]).to_csv("/content/metricas_quality.csv", index=False)

print("Arquivos salvos com sucesso!")
```
